```python
import json
import os

# 1. Định nghĩa đường dẫn
input_path = "dataset\\test.json"
output_path = "dataset\\test_2k.json" 

os.makedirs(os.path.dirname(output_path), exist_ok=True)

print(f"Đang đọc dữ liệu từ: {input_path}")

try:
    # Đọc toàn bộ dữ liệu từ file JSON
    with open(input_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
    print(f"Tổng số mẫu ban đầu: {len(data)}")

    # 2. Cắt dữ liệu: Chỉ giữ lại 20,000 mẫu đầu tiên
    NUM_SAMPLES_TO_KEEP = 2000
    
    if len(data) > NUM_SAMPLES_TO_KEEP:
        subset_data = data[:NUM_SAMPLES_TO_KEEP]
        print(f"Đã cắt dữ liệu, số mẫu mới: {len(subset_data)}")
    else:
        # Trường hợp file gốc có ít hơn 20k mẫu
        subset_data = data
        print("File gốc có ít hơn 20,000 mẫu, giữ lại toàn bộ.")

    # 3. Ghi dữ liệu đã cắt vào file JSON mới
    with open(output_path, 'w', encoding='utf-8') as f:
        json.dump(subset_data, f, ensure_ascii=False, indent=4)
        
    print(f"Đã ghi 20,000 mẫu vào file mới: {output_path}")
    
except FileNotFoundError:
    print(f"LỖI: Không tìm thấy file tại đường dẫn: {input_path}")
except Exception as e:
    print(f"Đã xảy ra lỗi trong quá trình xử lý: {e}")
```
